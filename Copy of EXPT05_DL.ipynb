```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import pandas as pd
dataframe = pd.read_csv("placementdataset.csv")
dataframe = dataframe.drop_duplicates()
X = dataframe.iloc[:,:2].values
Y = dataframe.iloc[:,2:3].values
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(X,Y,test_size=0.2)
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
weights_biases = []
for layer in model.layers:
  weights_biases.append(layer.get_weights())
# Print the weights and biases of each layer
for i, (weights, biases) in enumerate(weights_biases):
  print(f"Layer {i + 1} Weights:")
  print(weights)
  print(f"Layer {i + 1} Biases:")
  print(biases)
```
